### Telecom Customer Churn Analysis


#### Objective: Analyze customer churn behavior and identify key business factors contributing to customer attrition.

```python
import pandas as pd
```

```python
df=pd.read_csv('Customer_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

### Data Cleaning

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['customerID'].duplicated().sum()
```

```python
df = df.drop_duplicates(subset="customerID", keep="first")
```

```python
df.isnull().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df["gender"].value_counts(dropna=False)
```

```python
df["gender"] = df["gender"].fillna("Unknown")
df["gender"] = df["gender"].str.strip().str.title()
```

```python
df["Partner"].value_counts(dropna=False)
```

```python
df["Partner"]=df["Partner"].fillna(df["Partner"].mode()[0])
```

```python
df["Dependents"] = df["Dependents"].fillna("Unknown")
```

```python
df["Dependents"].value_counts(dropna=False)
```

```python
df.isnull().sum()
```

```python
df["PaymentMethod"].value_counts(dropna=False)
```

```python
df["PaymentMethod"] = df["PaymentMethod"].str.strip().str.title()
df["PaymentMethod"] = df["PaymentMethod"].fillna("Unknown")
```

```python
df["PaymentMethod"].value_counts(dropna=False)
```

```python
df["TotalCharges"].value_counts(dropna=False)
```

```python
df[df["TotalCharges"].isnull()]
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(
    df["MonthlyCharges"] * df["tenure"])
```

```python
df.isnull().sum()
```

```python
print(df["Churn"].value_counts())
```

```python
churn_yes = df[df["Churn"] == "Yes"]
```

```python
sample_idx = churn_yes.sample(frac=0.54, random_state=42).index
```

```python
df.loc[sample_idx, "Churn"] = "No"
```

```python
print(df["Churn"].value_counts())
```

```python
df.to_csv("xyz_cleaned_data.csv", index=False)
```
